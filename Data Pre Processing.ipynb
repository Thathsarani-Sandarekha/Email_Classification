```python
# Open the .names file and read the feature names into a list
with open('spambase.names', 'r') as f:
    reader = csv.reader(itertools.islice(f, 29, None))
    features = [row[0].replace("continuous.", "") for row in reader if row and row[0] != '']
    
# Open the .data file and read the data into a list of lists
with open('spambase.data', 'r') as f:
    reader = csv.reader(f)
    data = [row for row in reader]

# Replace the number index with the corresponding feature name in the first column of the data
for i in range(58):
    if i == 57:
        features[i] = "spam, non-spam classes"
        break
    else:
        features[i] = features[i + 1]
    
# Combine the feature names list and data list of lists into a single list of lists
combined = [features] + data

# Write the combined list of lists to a CSV file
with open('spam.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(combined)
```
